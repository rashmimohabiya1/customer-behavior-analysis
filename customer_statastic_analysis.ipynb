## *handling NAN values of Review Rating col*

```python
# filling missing values (NaN) in the Review Rating column based on each Category's median reviwe rating col.

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
df['Review Rating']
```

```python
# now Review Rating col is not containg any missing values
df['Review Rating'].isnull().sum()
```

## *Now convert columns name in snake case ie, lowecase alpha+underscore.*

```python
# Now convert columns name in snake case ie, lowecase alpha+underscodf.

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

## *Feature Engineering*

```python
# create col name age_group which divide age column into 4 equal-sized groups (quantiles) and label each group with a category like Young Adult, Adult, Middle-aged, Senior.

labels = ['Young Adult','Adult','Middle-Age','Senoir']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
df[['age','age_group']].head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# this col frequency_of_purchases contains text data , so we create col purchase_frequency_days col in which records are in numeric form so 
# its easy to further analysis

df['frequency_of_purchases'].unique()
```

```python
mapping_values = {
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(mapping_values)
df['purchase_frequency_days']
```
